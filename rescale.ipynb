```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# https://stackoverflow.com/questions/32946436/read-img-medical-image-without-header-in-python

dtype = np.uint8
shape = (200, 1024, 200)

with open('cubefiles/Macula200x200.img', 'rb') as f:
    data = np.fromfile(f, dtype)

img = data.reshape(shape)

slices = range(0, 200, 200 // 5)
for i, z in enumerate(slices):
    plt.subplot(2, 3, i + 1)
    plt.imshow(img[:, :, z], cmap='gray', aspect=1024 / 200)
    plt.title(f'z = {z}')

plt.tight_layout()
```

```python
pimg = Image.fromarray(img[:, :, 80])
pimg.size
```
